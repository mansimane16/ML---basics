```python
import pandas as pd 
df = pd.read_csv("homepricesohe.csv")
df
```

```python
dummies=pd.get_dummies(df.town)
dummies
```

```python
merged=pd.concat([df,dummies],axis=1)
```

```python
merged
```

```python
final = merged.drop(["town","west windsor"],axis=1)
```

```python
final
```

```python
X=final.drop("price",axis=1)
```

```python
X
```

```python
Y=final.price
```

```python
Y
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(X,Y)
```

```python
model.predict([[2600,0,0]])
```

```python
model.score(X,Y)
```
